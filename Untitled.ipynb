```python
# 시장 주중평균, 토요일, 일요일 매출합계
# 시장 일별 데이터 
def daily_data_cal(market_name, data): 
    is_market, market, weekday, satday, sunday = [], [], [], [], []

    for i in range(4):
        is_market.append((data['상권_코드_명'] == market_name)&(data['기준_분기_코드'] == i+1))
        #data[data.columns.difference(exception)]
        market.append(data[is_market[i]])

        weekday.append(int(market[i]['주중_매출_금액'].sum()/5))
        satday.append(market[i]['토요일_매출_금액'].sum())
        sunday.append(market[i]['일요일_매출_금액'].sum())
        
    return weekday, satday, sunday

def list_sub(a, b):
    a_sub_b = []
    for i in range(len(a)):
        a_sub_b.append(a[i] - b[i])
    return a_sub_b 

#비율 그래프 
def print_ratio(week,sat,sun):
    ratio_1 = [week[0], sat[0], sun[0]]
    labels_1 = ['Weekday', 'Saturday','Sunday']
    ratio_2 = [week[1], sat[1], sun[1]]
    labels_2 = ['Weekday', 'Saturday','Sunday']
    ratio_3 = [week[2], sat[2], sun[2]]
    labels_3 = ['Weekday', 'Saturday','Sunday']
    ratio_4 = [week[3], sat[3], sun[3]]
    labels_4 = ['Weekday', 'Saturday','Sunday']
    ratio = [sum(week), sum(sat), sum(sun)]
    labels = ['Weekday', 'Saturday','Sunday']

    # 비율
    plt.figure(figsize=(14, 14))
    plt.subplot(321)
    plt.pie(ratio_1,labels=labels_1, autopct='%.1f%%')
    plt.title('q1')

    plt.subplot(322)
    plt.pie(ratio_2,labels=labels_2, autopct='%.1f%%')
    plt.title('q2')

    plt.subplot(323)
    plt.pie(ratio_3,labels=labels_3, autopct='%.1f%%')
    plt.title('q3')

    plt.subplot(324)
    plt.pie(ratio_4,labels=labels_4, autopct='%.1f%%')
    plt.title('q4')

    plt.subplot(325)
    plt.pie(ratio,labels=labels, autopct='%.1f%%')
    plt.title('Total')

    plt.show()
    
# 분기 별 추이 그래프
def print_graph_quarter(week,sat,sun):
    x = ['1q','2q','3q','4q']
    plt.plot(x,week,label="Weekday", marker='o')
    plt.plot(x,sat, label="Saturday", marker='o')
    plt.plot(x,sun, label="Sunday", marker='o')
    plt.legend()
    plt.xlim(-0.5,3.5)
    plt.xlabel('Quarter')
    plt.ylabel('Sales (1M KRW)')
    plt.show()
    
# 전체 매출 막대 그래프
def print_graph_total(week,sat,sun):
    x = np.arange(3)
    y = [sum(week),sum(sat),sum(sun)]
    day = ['Weekday','Saturday','Sunday']
    colors = ['royalblue', 'darkorange', 'forestgreen']
    
    plt.bar(x,y, color=colors)
    plt.xticks(x, day)
    plt.xlim(-1,3)
    plt.xlabel('Total Sales')
    plt.ylabel('Sales (1M KRW)')
    plt.show()
    
# 백만원 단위 변경
def unit_mill_won(val):
    for i in range(4):
        val[i] = round(val[i]*(1/1000000))
    return val
```
